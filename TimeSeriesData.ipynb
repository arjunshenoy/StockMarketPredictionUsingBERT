This Notebook is part of stock market price prediction project and is used to extract the timeseries data such as stock price details, stock balance sheet details, and stock financial detials. We are making use of yahoo finanance to get the stock data and after transforming data according to our needs the dataset is finally exported as csv file which will be used in another notebook.

```python
pip install yfinance
# Import yfinance package
import yfinance as yf
import pandas as pd
from datetime import date
from numpy import exp
import os
from datetime import datetime
```

```python
#Set Ticker  here we are using the stock symbol of Tesla company its registered in NASDAQ as TSLA
ticker = 'TSLA'
stock = yf.Ticker(ticker)
# get stock info
stock.info
```

```python
# get historical market data
hist = stock.history(period="max")
```

```python

# Set the start and end date
start_date = '2018-01-01'
end_date = today = date.today()

# Get the data
df = stock.history(period ='max', rounding=True, start=start_date, end = end_date)
df.info()
df.head()
```

```python
# Print 5 rows
df.tail()
```

```python
#Getting balance sheet details for each year
df_balance_sheet=stock.get_balance_sheet()

#Doing a transpose of the balance sheet data
df_balance_sheet=df_balance_sheet.T
df_balance_sheet.info()
```

```python
#Getting the financials
df_financials = stock.get_financials()
#Taking the transpose
df_financials =  df_financials.T
for column in df_financials.columns:
  df_financials[column] = df_financials[column].astype(float, errors = 'raise')
df_financials.info()
```

```python

df_balance_sheet.iloc[:].head()
```

```python
# Merging both financial and balance sheet data to create a single dataset
df_fundemental = df_financials.join(df_balance_sheet.iloc[:], rsuffix='_duplicate')
df_fundemental.info()
```

```python
df_fundemental.head()
```

```python
df_fundemental.isnull().sum()

```

```python
#filling the null columns with 0 as the number of null columns are comparitively less
df_fundemental = df_fundemental.dropna(how='all', axis='columns')
df_fundemental.isnull().sum()
df_fundemental.fillna(0)
```

```python
df_fundemental.head()
```

```python
column_list = df_fundemental.columns
column_list
```

```python
#Combining the time series numerical data and the financials
def combineHistoricalAndFundemental(row):
  date_extracted = datetime.datetime.strptime(str(row.name), "%Y-%m-%d %H:%M:%S")
  year=date_extracted.year
  if year == 2022:
    year = 2021
  for column in column_list:
      row[column] = df_fundemental.loc[str(year)+'-12-31'][column][0]
  return row
df=df.apply(combineHistoricalAndFundemental, axis=1)
```

```python
# Labelling the target column comparing the close and open price values
def labelDataset(row):
  if row.Close-row.Open > 0:
    target='UP'
  else:
    target='DOWN'
  row['Target'] = target
  return row
df=df.apply(labelDataset, axis=1) 
```

```python
df.reset_index(inplace=True)
df.info()
df.tail()
```

```python
#Exporting the dataset
to_csv_timestamp = datetime.today().strftime('%Y%m%d_%H%M%S')
path = os.getcwd()
filename = 'StockMarketDataTesla.csv'
df.to_csv(filename, index = False)
```
